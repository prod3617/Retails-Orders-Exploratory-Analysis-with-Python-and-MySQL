```python
# import the data.
import pandas as pd
data = pd.read_csv(r'C:\Users\MANJEET KUMAR\Downloads\orders.csv', na_values = ['Not Available', 'unknown'])
data['Ship Mode'].unique()
```

```python
# Rename the column by converting them into lower case and replace space with underscore.
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')
data.columns
#Another way to do that is:
#data.rename(columns = {'Order id': 'order_id', 'City': 'city', 'Order Date': 'order_date'}) 
```

```python
data.head()
```

```python
# Derive new columns like discount, sell price, profit.
data['discount'] = data['list_price'] * data['discount_percent']/100
data['sale_price'] = data['list_price'] - data['discount']
data['profit'] = data['sale_price'] - data['cost_price']
data.head()
```

```python
# Convert Date column datatype from object to datetime
data['order_date'] = pd.to_datetime(data['order_date'], format = '%Y-%m-%d')
#data.dtypes
```

```python
# Drop the columns that are not required now.
data.drop(columns = ['cost_price', 'list_price', 'discount_percent'], inplace = True)
```
